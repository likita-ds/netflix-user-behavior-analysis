```python
#visualization code
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#First visualization (Categorical column)
df['gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top Countries')
plt.show()
```

```python
plt.hist(df['age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='age', y='avg_watch_time_minutes', data=df)
plt.title('Age vs Watch Time')
plt.show()
```

```python
sns.boxplot(x='gender', y='avg_watch_time_minutes', data=df)
plt.title('Watch Time by Gender')
plt.show()
```

```python
corr = df.corr(numeric_only=True)

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr, annot=True)
plt.show()
```

```python
#GroupBy Analysis
df.groupby('country')['avg_watch_time_minutes'].mean().sort_values(ascending=False)
```

```python
df.groupby('country')['avg_watch_time_minutes'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top Countries by Watch Time')
plt.show()
```

```python
df.groupby('gender')['avg_watch_time_minutes'].mean()
```

```python
sns.barplot(x='gender', y='avg_watch_time_minutes', data=df)
plt.title('Watch Time by Gender')
plt.show()
```

```python
df.sort_values(by='avg_watch_time_minutes', ascending=False).head()
```

```python
df['watch_category'] = df['avg_watch_time_minutes'].apply(lambda x: 'High' if x > 5 else 'Low')
```
